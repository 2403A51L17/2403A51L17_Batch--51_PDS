```python
import pandas as pd
import matplotlib.pyplot as plt

budget_data = {
    'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Operations'],
    'Budget': [50000, 150000, 120000, 80000, 100000]
}
df_budget = pd.DataFrame(budget_data)

expenses_data = {
    'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Operations'],
    'Actual_Expenses': [52000, 140000, 125000, 90000, 95000]
}
df_expenses = pd.DataFrame(expenses_data)

df_merged = pd.merge(df_budget, df_expenses, on='Department')
df_merged['Variance'] = df_merged['Actual_Expenses'] - df_merged['Budget']
df_merged['Exceed_Budget'] = df_merged['Variance'] > 0
df_sorted = df_merged.sort_values(by='Variance', ascending=False)

print(df_sorted)

plt.figure(figsize=(8, 5))
bar_width = 0.35
index = range(len(df_sorted))
plt.bar(index, df_sorted['Budget'], bar_width, label='Budget')
plt.bar([i + bar_width for i in index], df_sorted['Actual_Expenses'], bar_width, label='Actual Expenses')
plt.xlabel('Department')
plt.ylabel('Amount')
plt.title('Budget vs Actual Expenses by Department')
plt.xticks([i + bar_width/2 for i in index], df_sorted['Department'])
plt.legend()
plt.tight_layout()
plt.show()
```
