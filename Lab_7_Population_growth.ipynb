```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('world_population.csv')

country = 'India'
country_data = df[df['Country/Territory'] == country]

year_cols = ['2022 Population', '2020 Population', '2015 Population', '2010 Population',
             '2000 Population', '1990 Population', '1980 Population', '1970 Population']

melted = country_data.melt(id_vars=['Country/Territory'], value_vars=year_cols,
                           var_name='Year', value_name='Population')

melted['Year'] = melted['Year'].str.extract('(\d{4})').astype(int)

melted = melted.sort_values('Year')

plt.figure(figsize=(8,5))
plt.plot(melted['Year'], melted['Population'], marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population Growth of {country} Over Years')
plt.grid(True)
plt.tight_layout()
plt.show()
```
