```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sales_category.csv")

plt.figure(figsize=(8,6))
plt.bar(df['Category'], df['Sales'],
        color=['skyblue','orange','green','red','purple'])

plt.xlabel("Product Categories")
plt.ylabel("Sales Amount")
plt.title("Sales Distribution Across Product Categories")
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

best_category = df.loc[df['Sales'].idxmax(), 'Category']
print(f"Best-Selling Category: {best_category}")
```
