```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, BaggingClassifier, IsolationForest
from sklearn.svm import SVC
from sklearn.cluster import KMeans, DBSCAN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# ===============================
# 🔹 Step A: Feature Engineering for Phishing (Keep this function)
# ===============================
def add_phishing_features(df, url_column="url"):
    if url_column in df.columns:
        df["url_length"] = df[url_column].apply(lambda x: len(str(x)))
        df["count_dot"] = df[url_column].apply(lambda x: str(x).count('.'))
        df["count_at"] = df[url_column].apply(lambda x: str(x).count('@'))
        df["count_hyphen"] = df[url_column].apply(lambda x: str(x).count('-'))
        df["count_slash"] = df[url_column].apply(lambda x: str(x).count('/'))
        df["has_https"] = df[url_column].apply(lambda x: 1 if "https" in str(x).lower() else 0)
        df["digit_count"] = df[url_column].apply(lambda x: sum(c.isdigit() for c in str(x)))
        df["letter_count"] = df[url_column].apply(lambda x: sum(c.isalpha() for c in str(x)))
        df.drop(columns=[url_column], inplace=True, errors="ignore")
    return df

# ===============================
# 🔹 Step D: Utility Function for Metrics (Keep this function)
# ===============================
def get_metrics(y_true, y_pred, name):
    acc = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, average='weighted')
    pre = precision_score(y_true, y_pred, average='weighted')
    rec = recall_score(y_true, y_pred, average='weighted')
    # print(f"{name} -> Accuracy: {acc:.3f}, F1: {f1:.3f}, Precision: {pre:.3f}, Recall: {rec:.3f}")
    return {"model": name, "accuracy": acc * 100, "f1": f1 * 100, "precision": pre * 100, "recall": rec * 100}

# ===============================
# 🔹 Step B: Load Datasets and Prepare for Loop
# ===============================
# Load all datasets into a dictionary after feature engineering
datasets = {
    "D1": add_phishing_features(pd.read_csv("/content/dataset1.csv", encoding="latin1")),
    "D2": add_phishing_features(pd.read_csv("/content/dataset2.csv", encoding="latin1")),
    "D3": add_phishing_features(pd.read_csv("/content/dataset3.csv", encoding="latin1")),
    "D4": add_phishing_features(pd.read_csv("/content/dataset4.csv", encoding="latin1")),
    "D5": add_phishing_features(pd.read_csv("/content/dataset5.csv", encoding="latin1", on_bad_lines='skip')),
    "D6": add_phishing_features(pd.read_csv("/content/dataset6.csv", encoding="latin1")),
}

all_results = []
scaler = StandardScaler()

# ===============================
# 🔹 Step C: Main Processing Loop
# ===============================
for name, df in datasets.items():
    print(f"\n--- Processing Dataset: {name} (Shape: {df.shape}) ---")

    # 1. Preprocessing and Splitting
    # Handle 'label' column and ensure it's binary
    if 'label' in df.columns:
        # Convert 'label' column to numeric, coercing errors to NaN
        df['label'] = pd.to_numeric(df['label'], errors='coerce')
        # Drop rows where 'label' is NaN
        df = df.dropna(subset=['label'])
        # Convert 'label' to integer type
        df['label'] = df['label'].astype(int)
        # Map unique values to 0 and 1 if there are more than 2 unique values
        if df['label'].nunique() > 2:
            unique_labels = df['label'].unique()
            # Assuming the two most frequent values are the binary labels
            top_two_labels = df['label'].value_counts().nlargest(2).index.tolist()
            label_map = {label: 1 if label == top_two_labels[0] else 0 for label in unique_labels}
            df['label'] = df['label'].map(label_map)
            print(f"Converted 'label' column in {name} to binary. New unique values: {df['label'].unique()}")
    else:
        # If 'label' column does not exist, create dummy binary labels
        df['label'] = np.random.randint(0, 2, size=len(df))
        print(f"Created dummy 'label' column for {name}.")


    # Convert all columns to numeric types, coercing errors
    for col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    df = df.dropna()

    if df.shape[0] == 0:
        print(f"Skipping {name} as it is empty after preprocessing.")
        continue

    X = df.drop(columns=["label"], errors="ignore")
    y = df["label"]

    # Scale features
    X_scaled = scaler.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

    # 2. Supervised & Ensemble Models (Train on X_train, Test on X_test)

    # Logistic Regression
    log = LogisticRegression(max_iter=500).fit(X_train, y_train)
    all_results.append(get_metrics(y_test, log.predict(X_test), f"LR ({name})"))

    # Random Forest
    rf = RandomForestClassifier().fit(X_train, y_train)
    all_results.append(get_metrics(y_test, rf.predict(X_test), f"RF ({name})"))

    # Gradient Boosting
    gb = GradientBoostingClassifier().fit(X_train, y_train)
    all_results.append(get_metrics(y_test, gb.predict(X_test), f"GB ({name})"))

    # Deep Learning ANN
    ann = Sequential([
        Dense(64, input_dim=X_train.shape[1], activation='relu'),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    ann.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    ann.fit(X_train, y_train, epochs=10, batch_size=32, verbose=0)
    y_ann = (ann.predict(X_test, verbose=0) > 0.5).astype(int)
    all_results.append(get_metrics(y_test, y_ann, f"ANN ({name})"))

    # 3. Unsupervised Models (Fit on X_scaled, Evaluate on full y)

    # KMeans
    kmeans = KMeans(n_clusters=2, random_state=42, n_init='auto').fit(X_scaled)
    all_results.append(get_metrics(y, kmeans.labels_, f"KMeans ({name})"))

    # Isolation Forest
    iso = IsolationForest(random_state=42).fit(X_scaled)
    iso_labels = np.where(iso.predict(X_scaled) == -1, 0, 1) # -1 is anomaly (phishing), map to 0/1
    all_results.append(get_metrics(y, iso_labels, f"IsoForest ({name})"))

# ===============================
# 🔹 Step I: Consolidated Evaluation
# ===============================
results_df = pd.DataFrame(all_results)
print("\n--- Consolidated Results Across All Datasets ---")
print(results_df.sort_values(by=["accuracy"], ascending=False).head(20))

# ===============================
# 🔹 Step J: Individual Model Plots
# ===============================
print("\n--- Individual Model Performance Plots ---")
for index, row in results_df.iterrows():
    model_name = row['model']
    accuracy = row['accuracy']
    f1 = row['f1']

    plt.figure(figsize=(6, 4))
    sns.barplot(x=['Accuracy', 'F1 Score'], y=[accuracy, f1], palette='viridis')
    plt.title(f'Performance Metrics for {model_name}')
    plt.ylabel('Score (%)')
    plt.ylim(0, 100)
    # Add percentage labels on top of bars
    for i, v in enumerate([accuracy, f1]):
        plt.text(i, v + 1, f"{v:.2f}%", ha='center')
    plt.show()

# ===============================
# 🔹 Step K: Identify Best Algorithm
# ===============================
best_model_accuracy = results_df.sort_values(by=["accuracy"], ascending=False).iloc[0]
best_model_f1 = results_df.sort_values(by=["f1"], ascending=False).iloc[0]

print("\n🏆 Best Model Based on Accuracy:")
print(best_model_accuracy)

print("\n🏆 Best Model Based on F1 Score:")
print(best_model_f1)
```
