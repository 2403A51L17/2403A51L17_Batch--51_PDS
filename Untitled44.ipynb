```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student_marks.csv")

plt.figure(figsize=(8,6))
plt.bar(df['Subject'], df['Average_Marks'], color=['skyblue','orange','green','red','purple'])

plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.title("Average Marks of Students in Different Subjects")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

easiest_subject = df.loc[df['Average_Marks'].idxmax(), 'Subject']
toughest_subject = df.loc[df['Average_Marks'].idxmin(), 'Subject']

print(f"Easiest Subject (Highest Avg Marks): {easiest_subject}")
print(f"Toughest Subject (Lowest Avg Marks): {toughest_subject}")
```
